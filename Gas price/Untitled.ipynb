```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gas_prices.csv')
```

```python
plt.figure(figsize=(10,5))
plt.title('Variation Over the Years(USA)')
plt.plot(df['Year'],df['USA'], marker="o", linestyle="-", color="b", label="EUA")
plt.xlabel("Ano")
plt.ylabel("Preço da Gasolina (USD)")
plt.xticks(df['Year'][::2])
plt.legend()
plt.grid(True)
plt.show()           
```

```python
colunas_lista = ['Australia', 'Canada', 'France', 'Germany', 'Italy', 'Japan', 'Mexico', 'South Korea', 'UK', 'USA']
plt.figure(figsize=(10,8))
plt.title('Comparação Entre os Países')
for pais in df:
    if pais in colunas_lista:
        plt.plot(df['Year'],df[pais], label=[pais] ,marker='.')
plt.xticks(df['Year'][::2])
plt.xlabel('Ano')
plt.ylabel('US$')
plt.legend(ncol=2)       
plt.show()
```

```python
preços=df.set_index('Year')
preço_mais_alto = preços.idxmax(axis=1)
preco_mais_baixo= preços.idxmin(axis=1)
extremos_comparados = pd.DataFrame({"Mais Caro":preço_mais_alto, "Mais Baixo":preco_mais_baixo})
```

```python
contagem_mais_alto = extremos_comparados['Mais Caro'].value_counts()
contagem_mais_baixo = extremos_comparados['Mais Baixo'].value_counts()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
contagem_mais_alto.plot(kind='bar',color='red', alpha=0.7)
plt.title("Número de vezes como país mais caro")
plt.xticks(rotation=45)
plt.ylabel("Frequência")

plt.subplot(1,2,2)
contagem_mais_baixo.plot(kind='bar',color='green',)
plt.title("Número de vezes como país mais baixo")
plt.xticks(rotation=45)
plt.ylabel("Frequência")

plt.tight_layout()
plt.show()
```
